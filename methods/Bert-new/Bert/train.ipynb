```python
import logging

# 创建日志器
logger = logging.getLogger('no_warning_logger')
logger.setLevel(logging.DEBUG)  # 设置日志器级别

# 创建处理器（输出到标准输出）
console_handler = logging.StreamHandler()
console_handler.setLevel(logging.DEBUG)  # 控制台输出级别

# 定义日志格式
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
console_handler.setFormatter(formatter)

# 添加处理器到日志器
logger.addHandler(console_handler)

# 自定义过滤器，排除 WARNING 及以上级别
class NoWarningFilter(logging.Filter):
    def filter(self, record):
        return record.levelno < logging.WARNING

# 给处理器添加过滤器
console_handler.addFilter(NoWarningFilter())
```
